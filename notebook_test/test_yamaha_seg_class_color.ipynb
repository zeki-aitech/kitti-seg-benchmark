```python
def plot_color(rgb):
    """
    Plots a square of the given RGB color.

    Parameters:
        rgb (tuple or list): A tuple or list containing the RGB values (R, G, B) in the range [0, 255].
    """
    # Normalize RGB values to [0, 1] for matplotlib
    normalized_rgb = [value / 255.0 for value in rgb]
    
    # Create a square of the color
    color_square = np.zeros((100, 100, 3))
    color_square[:, :] = normalized_rgb
    
    # Plot the color
    plt.imshow(color_square)
    plt.axis('off')  # Turn off axes
    plt.title(f"Color: RGB{rgb}")
    plt.show()
```

```python
for color in unique_colors:
    # Convert the color to a tuple
    color_tuple = tuple(color)
    
    # Plot the color
    plot_color(color_tuple)
```
